```python
import pandas as pd
import os as os
def concat_user_files(path):
    data = pd.DataFrame()
    file_list = os.listdir(path)
    file_list.sort()
    
    for file in file_list:
        df = pd.read_csv(path + '/' + file)
        data = pd.concat(
            [data, df],
            axis=0,
            ignore_index=True
        )
    data = data.drop_duplicates()
    return data
```
